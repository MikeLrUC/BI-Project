```python
import matplotlib.pyplot as plt
```

```python
def worldwide_plot(feature):
    plt.figure(figsize=(10,10))
    plt.scatter(df["lng"], df["lat"], c=df[feature])
    plt.colorbar()
    plt.xlim([-180, 180])
    plt.ylim([-90, 90])
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.title(f"{feature.capitalize()} over the world");
```

```python
worldwide_plot("rating")
```

```python
worldwide_plot("alcohol")
```

```python
worldwide_plot("price")
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["alcohol"], df["price"], c=df["rating"])
plt.colorbar()
plt.xlabel("Alcohol")
plt.ylabel("Price")
plt.title(f"Rating over the alcohol and price");
```

```python
plt.figure(figsize=(10,10))
plt.scatter(df["alcohol"], df["age"], c=df["rating"])
plt.colorbar()
plt.xlabel("Alcohol")
plt.ylabel("Age")
plt.title(f"Rating over the alcohol and age");
```
